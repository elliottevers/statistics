# Foundations for inference - Confidence intervals

```r
set.seed(9102016)
```

```r
library(statsr)
library(dplyr)
library(ggplot2)
```

```r
data(ames)
```

```r
n <- 60
samp <- sample_n(ames, n)
```

# Q: Describe the distribution of homes in your sample. What would you say is the “typical” size within your sample? Also state precisely what you interpreted “typical” to mean.

```r
ggplot(
    data=samp,
    aes(x=area)
) + geom_histogram(
    binwidth=100
)
```

```r
summarise(
    samp,
    mean=mean(area)
)
```

# A: The distribution of area for this example is right-skewed.  The "typical" size of a house is 1502.15 feet^2, if we're using the mean to measure that notion.

# Q: True or False: My distribution should be similar to others’ distributions who also collect random samples from this population, but it is likely not exactly the same since it’s a random sample.
- True.
- False.

# A: True

```r
z_star_95 <- qnorm(0.975)
z_star_95
```

```r
interval <- samp %>%
  summarise(lower = mean(area) - z_star_95 * (sd(area) / sqrt(n)),
            upper = mean(area) + z_star_95 * (sd(area) / sqrt(n)))
interval
```

# Q: For the confidence interval to be valid, the sample mean must be normally distributed and have standard error $\frac{s}{\sqrt{n}}$. Which of the following is not a condition needed for this to be true?
- The sample is random.
- The sample size, 60, is less than 10% of all houses.
- The sample distribution must be nearly normal.

# A: The sample size, 60, is less than 10% of all houses.

# Q: What does “95% confidence” mean?
- 95% of the time the true average area of houses in Ames, Iowa, will be in this interval.
- 95% of random samples of size 60 will yield confidence intervals that contain the true average area of houses in Ames, Iowa.
- 95% of the houses in Ames have an area in this interval.
- 95% confident that the sample mean is in this interval.

# A: 95% of random samples of size 60 will yield confidence intervals that contain the true average area of houses in Ames, Iowa.

```r
params <- ames %>%
  summarise(
      mu = mean(area)
  )
```

# Q: Does your confidence interval capture the true average size of houses in Ames?

```r
mu <- params[1][1]
lbound <- interval[1]
ubound <- interval[2]
lbound < mu && mu < ubound
```

# A: Yes

# Q: What proportion of 95% confidence intervals would you expect to capture the true population mean?

# A: 95/100

```r
ci <- ames %>%
        rep_sample_n(size = n, reps = 50, replace = TRUE) %>%
        summarise(lower = mean(area) - z_star_95 * (sd(area) / sqrt(n)),
                  upper = mean(area) + z_star_95 * (sd(area) / sqrt(n)))
```

```r
ci
```

```r
ci <- ci %>%
  mutate(
      capture_mu = ifelse(lower < params$mu & upper > params$mu, "yes", "no")
  )
```

```r
ci
```

```r
ci_data <- data.frame(
    ci_id = c(1:50, 1:50),
    ci_bounds = c(ci$lower, ci$upper),
    capture_mu = c(ci$capture_mu, ci$capture_mu)
)
```

```r
ggplot(
    data = ci_data,
    aes(
        x = ci_bounds,
        y = ci_id, 
        group = ci_id,
        color = capture_mu
    )
) +
  geom_point(size = 2) +  # add points at the ends, size = 2
  geom_line() +           # connect with lines
  geom_vline(xintercept = params$mu, color = "darkgray") # draw vertical line
```

# Q: What proportion of your confidence intervals include the true population mean? Is this proportion exactly equal to the confidence level? If not, explain why.

# A: Only 94% trapped the true mean.  Well, it *can't* be equal to it because we only have 50 samples, and would need *47.5* of the intervals to trap the mean, which is illogical.

# Q: What is the appropriate critical value for a 99% confidence level?
- 0.01
- 0.99
- 1.96
- 2.33
- 2.58

```r
z_star_99 <- qnorm(0.995)
z_star_99
```

# A: 2.58

# Q: Calculate 50 confidence intervals at the 99% confidence level. You do not need to obtain new samples, simply calculate new intervals based on the 95% confidence interval endpoints you had already collected. Plot all intervals and calculate the proportion of intervals that include the true population mean.

```r
ci_99 <- ames %>%
        rep_sample_n(size = n, reps = 50, replace = TRUE) %>%
        summarise(lower = mean(area) - z_star_99 * (sd(area) / sqrt(n)),
                  upper = mean(area) + z_star_99 * (sd(area) / sqrt(n)))

ci_99 <- ci_99 %>%
  mutate(
      capture_mu = ifelse(lower < params$mu & upper > params$mu, "yes", "no")
  )

ci_data_99 <- data.frame(
    ci_id = c(1:50, 1:50),
    ci_bounds = c(ci_99$lower, ci_99$upper),
    capture_mu = c(ci_99$capture_mu, ci_99$capture_mu)
)

ggplot(
    data = ci_data_99,
    aes(
        x = ci_bounds,
        y = ci_id, 
        group = ci_id,
        color = capture_mu
    )
) +
  geom_point(size = 2) +  # add points at the ends, size = 2
  geom_line() +           # connect with lines
  geom_vline(xintercept = params$mu, color = "darkgray") # draw vertical line
```

# A: See Above
