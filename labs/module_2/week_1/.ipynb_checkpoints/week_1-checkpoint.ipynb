# Foundations for inference - Sampling distributions

```r
library(statsr)
library(dplyr)
library(shiny)
library(ggplot2)
```

```r
data(ames)
```

```r
ggplot(data = ames, aes(x = area)) +
  geom_histogram(binwidth = 250)
```

```r
ames %>%
  summarise(mu = mean(area), pop_med = median(area), 
            sigma = sd(area), pop_iqr = IQR(area),
            pop_min = min(area), pop_max = max(area),
            pop_q1 = quantile(area, 0.25),  # first quartile, 25th percentile
            pop_q3 = quantile(area, 0.75))  # third quartile, 75th percentile
```

# Q: Which of the following is false?
- The distribution of areas of houses in Ames is unimodal and right-skewed.
- 50% of houses in Ames are smaller than 1,499.69 square feet.
- The middle 50% of the houses range between approximately 1,126 square feet and 1,742.7 square feet.
- The IQR is approximately 616.7 square feet.
- The smallest house is 334 square feet and the largest is 5,642 square feet.

# A: 50% of houses in Ames are smaller than 1,499.69 square feet

```r
samp1 <- ames %>%
  sample_n(size = 50)
```

# Q: Describe the distribution of this sample? How does it compare to the distribution of the population?

```r
ggplot(
    data = samp1,
    aes(x = area)
) + geom_histogram(
    binwidth = 250
)
```

# A:  It's also right-skewed, with a sample mean similar to the population mean

```r
samp1 %>%
  summarise(x_bar = mean(area))
```

# Q: Suppose we took two more samples, one of size 100 and one of size 1000. Which would you think would provide a more accurate estimate of the population mean?
- Sample size of 50.
- Sample size of 100.
- Sample size of 1000.

# A: Sample size of 1000

```r
ames %>%
  sample_n(size = 50) %>%
  summarise(x_bar = mean(area))
```

```r
sample_means50 <- ames %>%
    rep_sample_n(
        size = 50,
        reps = 15000,
        replace = TRUE
    ) %>%
        summarise(
            x_bar = mean(area)
        )

ggplot(
    data = sample_means50,
    aes(x = x_bar)
) +
geom_histogram(
    binwidth = 20
)
```

# Q: How many elements are there in sample_means50? Describe the sampling distribution, and be sure to specifically note its center. Make sure to include a plot of the distribution in your answer.

```r
sample_means50 %>%
    summarise(
        sampling_mean=mean(x_bar)
    )
```

# A: There are 15,000 elements in sample_means50.  The distribution is very close to a Gaussian, and the mean of the sample means is 1499.297

# A: 178093.7

# Q: Since you have access to the population, simulate the sampling distribution for x¯price by taking 5000 samples from the population of size 50 and computing 5000 sample means. Store these means in a vector called sample_means50. Plot the data, then describe the shape of this sampling distribution. Based on this sampling distribution, what would you guess the mean home price of the population to be?

```r
sample_means50 <- ames %>% rep_sample_n(
    reps = 5000,
    size = 50,
    replace = TRUE
) %>% summarise(
    x_bar=mean(price)
)

ggplot(
    data=sample_means50,
    aes(x=x_bar)
) + geom_histogram(
    binwidth = 1000
)
```

```r
summarise(
    sample_means50,
    sampling_mean_50=mean(x_bar)
)
```

# A: 181232.8

# Q: Change your sample size from 50 to 150, then compute the sampling distribution using the same method as above, and store these means in a new vector called sample_means150. Describe the shape of this sampling distribution, and compare it to the sampling distribution for a sample size of 50. Based on this sampling distribution, what would you guess to be the mean sale price of homes in Ames?

```r
sample_means150 <- ames %>% rep_sample_n(
    reps = 5000,
    size = 150,
    replace = TRUE
) %>% summarise(
    x_bar=mean(price)
)

ggplot(
    data=sample_means50,
    aes(x=x_bar)
) + geom_histogram(
    binwidth = 1000
)
```

```r
summarise(
    sample_means150,
    sampling_mean_150=mean(x_bar)
)
```

# A: The interquatile range of this distribution of sample averages is smaller than the former.  My new estimate for the true average is 180638.8.

# Q: Since you have access to the population, simulate the sampling distribution for x¯price by taking 2000 samples from the population of size 15 and computing 2000 sample means. Store these means in a vector called sample_means15. Plot the data, then describe the shape of this sampling distribution. Based on this sampling distribution, what would you guess the mean home price of the population to be? Finally, calculate and report the population mean.

```r
sample_means15 <- ames %>%
    rep_sample_n(
        reps = 2000,
        size = 15,
        replace = TRUE
    ) %>%
        summarise(
            x_bar=mean(price)
        )

ggplot(
    data=sample_means15,
    aes(x=x_bar)
) + geom_histogram(
    binwidth=4000
)
```

```r
summarise(
    sample_means15,
    samplling_average_50=mean(x_bar)
)
```

# A: The distribution is nearly normal - I would expect the population mean of `price` to be slightly higher than the sampling mean of `179910.6`.
