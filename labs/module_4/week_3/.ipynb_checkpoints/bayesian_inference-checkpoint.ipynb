```r
library(PairedData)
library(tidyverse)
library(statsr)
```

```r
data(nc)
```

```r
summary(nc)
```

```r
ggplot(bdata = nc, aes(x = gained)) +
  geom_histogram(binwidth = 5)
```

```r
nc_fullterm = filter(nc, premie == 'full term')
summary(nc_fullterm)
```

```r
bayes_inference(
    y = weight,
    data = nc_fullterm, 
    statistic = "mean",
    type = "ci",  
    prior_family = "JZS",
    mu_0 = 7.7,
    rscale = 1,
    method = "simulation",
    cred_level = 0.95
)
```

```r
bayes_inference(
    y = weight,
    data = nc_fullterm, 
    statistic = "mean",
    type = "ht",
    alternative = "twosided",
    prior_family = "JZS",
    mu_0 = 7.7,
    rscale = 1,
    method = "theoretical",
    show_plot=FALSE
)
```

```r
weight_post = bayes_inference(
    y = weight,
    data = nc_fullterm, 
    statistic = "mean",
    type = "ci",  
    prior_family = "JZS",
    mu_0 = 7.7,
    rscale = 1,
    method = "simulation",
    cred_level = 0.95
)
```

```r
names(weight_post)
```

```r
samples = as.data.frame(weight_post$samples)
nsim = nrow(samples)
samples = mutate(samples, y_pred = rnorm(nsim, mu, sqrt(sig2)))
```

```r
ggplot(data = samples, aes(x = y_pred)) + 
  geom_histogram(aes(y = ..density..), bins = 100) +
  geom_density() + 
  xlab(expression(y[new]))
```

```r
dplyr::select(samples, mu, y_pred) %>%
  map(quantile, probs=c(0.025, 0.50, 0.975))
```
