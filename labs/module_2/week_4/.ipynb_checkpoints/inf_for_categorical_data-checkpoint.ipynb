# Inference for categorical data

```r
library(statsr)
library(dplyr)
library(ggplot2)
```

```r
data(atheism)
```

```r
str(atheism)
```

```r
us12 <- atheism %>%
  filter(nationality == "United States" , atheism$year == "2012")
```

```r
inference(
    y = response,
    data = us12,
    statistic = "proportion",
    type = "ci",
    method = "theoretical",
    success = "atheist"
)
```
